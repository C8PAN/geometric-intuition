# Co-Linearity Example


```python
import numpy as np
import pandas as pd
import altair as alt
```

```python
def salary(dash100_speed):
    q = 1.0 / dash100_speed * np.random.normal(1.0, 0.1, size=N)
    q = (q - q.min()) / (q.max() - q.min())
    q = q ** 1.0
    return q * 1.8 + 0.20
```

```python
N = 300
np.random.seed(123)
spd = np.random.normal(20,2.5, size=N)
d = pd.DataFrame(data={
    '100m dash (s)': spd,
    '100yd dash (s)': spd * np.random.normal(1.1, 0.03, size=N),
    '100m dash (min)': spd/60,
    '200m dash (s)': spd * 2.2 * np.random.normal(1.0, 0.2, size=N) + 10,
    'salary': salary(spd),
})
```

```python
d[:10]
```

```python
alt.Chart(data=d,
          height=700,
          width=1200,
).configure_axis(
    domain=False,
    labelFontSize=26,
    titleFontSize=33,
).mark_point(
    size=100,
    opacity=1.0,
    filled=True,
).encode(
    alt.X('100m dash (s)',
          scale=alt.Scale(zero=False),
         ),
    alt.Y('200m dash (s)',
          scale=alt.Scale(zero=False),
         ),
    color=alt.Color('salary', scale=alt.Scale(scheme='plasma'))
)
```
